```python
#libreria para extraer coordenadas a partir de una dirección
from opencage.geocoder import OpenCageGeocode
import numpy as np
import pandas as pd
```

```python
#data ubigeos extraida del MEF
DataUbigeos = pd.read_csv('ubigeos-completos.csv')
DataUbigeos
```

```python
#cuando creas un perfil en Open Cage Data tienes la opccion de obtener un key free


key = 'e5d83df9c7cc4dc897474c757aeb82f3' #AQUI DEBES PONER EL KEY
geocoder = OpenCageGeocode(key)
```

```python
list_lat = []   # create empty lists

list_long = []


#extraer con loop las coordenadas de DF anterior
#trata de probar antes sextraer ccoordinadas sin este loop 
# porque el key te permite un límite de request

for index, row in DataUbigeos.iterrows(): # iterate over rows in dataframe

    
    City = row['DISTRITO']
    State = row['DEPARTAMENTO']  
        
    query = str(City)+','+ str(State)
    
    ## ex: geocoder.geocode("Krishnamurti Puram, Mysore, India")
    results = geocoder.geocode(query)   
    
    lat = results[0]['geometry']['lat']
    longi = results[0]['geometry']['lng']

    list_lat.append(lat)
    list_long.append(longi)


# create new columns from lists    

DataUbigeos['lat'] = list_lat   

DataUbigeos['lon'] = list_long

```
